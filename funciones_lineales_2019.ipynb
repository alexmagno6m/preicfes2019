# Funcion lineal

```python
import numpy as np
from bokeh.layouts import column, row
from bokeh.models import CustomJS, ColumnDataSource, Slider
from bokeh.plotting import Figure, output_file, show
from bokeh.io import output_notebook, show, reset_output
reset_output()
output_notebook()
x = np.arange(-10,10,1)
#x = [x*0.005 for x in range(0, 200)]
y = x

source = ColumnDataSource(data=dict(x=x, y=y))

plot = Figure(plot_width=450, plot_height=450, title="funcion  f(x)=ax+b")
plot.line('x', 'y', source=source, line_width=3, line_alpha=0.6)


plot.xaxis.axis_line_width = 3
plot.xaxis.axis_line_color = "red"
plot.xaxis.fixed_location = 0
plot.xaxis.axis_label = 'x'
plot.yaxis.axis_line_width = 3
plot.yaxis.axis_line_color = "red"
plot.yaxis.fixed_location = 0

slider = Slider(start=-10, end=10, value=0, step=1, title="intercepto (b)")
pendiente = Slider(start=-5, end=5, value=1, step=1, title="pendiente (a, numero que multiplica a x)")

callback = CustomJS(args=dict(source=source, sl1=slider, pen=pendiente), code="""
    var data = source.data;
    var sl1 = sl1.value;
    var pen = pen.value;
    var x = data['x']
    var y = data['y']
    for (var i = 0; i < x.length; i++) {
        //y[i] = Math.pow(x[i], f)
        y[i]= x[i]*pen + sl1;
    }
    source.change.emit();
""")


slider.js_on_change('value', callback)
pendiente.js_on_change('value', callback)

layout = column( plot, slider, pendiente)

show(layout)
```
