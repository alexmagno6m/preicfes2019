## Números decimales: 
Son todos aquellos números que no son enteros y pueden ser racionales (el cociente de una división exacta o periódica: 5÷2,1÷3) o irracionales (una aproximación a un valor, no periódicos e infinitos, por ejemplo: √2,π).  La coma (o el punto en algunos casos) separando la parte entera de la decimal los diferencia del resto de números.

### Las sumas y restas
Se operan como si fueran enteros alineando las comas

__Ejemplos__

$3.5 + 7.8$

$6.1 - 8.3$

Respuesta

```python
def operaciones():
    print('a) ',3.5+7.8)
    print('b) ',6.1-8.3)
operaciones()
```

### Las multiplicaciones

Al final en el resultado se cuentan de derecha a izquierda el total de posiciones decimales que tenian las cantidades a multiplicar


__Ejemplo__

$4.6*2.7$

Respuesta

```python
def multiplicacion():
    print(4.6*2.7)
multiplicacion()
```

### Problema:
Eva sigue un régimen de adelgazamiento y no puede pasar en cada comida de 600 calorías. Ayer almorzó : 125 g de pan, 140 g de espárragos, 45 g de queso y una manzana de 130 g. Si 1 g de pan da 3.3 calorías, 1 g de espárragos 0.32, 1 g de queso 1.2 y 1 g de manzana 0.52. ¿Respetó Eva su régimen?


__Consejo__ Debe multiplicar la cantidad de gramos por la cantidad de calorias para cada tipo de alimento

Pan $125*3.3$

Esparagos $140*0.32$

Queso $45*1.2$

Manzana $130*0.52$

Sume cada uno de los resultados

__Respuesta__

```python
alimentos = ['pan', 'esparragos', 'queso', 'manzana']
gramos = [125, 140, 45, 130]
calorias = [3.3, 0.32, 1.2, .52]
def problema():
    b = 0
    for i in range (len(alimentos)):
        a = gramos[i]*calorias[i]
        print('el total de {} es {}'.format(alimentos[i], a))
        b = a + b
    
    return b
total = problema()
print('el total de calorias fue', total)
if total <= 600:
    print('Eva respeto su regimen de comidas')
else:
    print('Eva no respeto su regimende comidas')        
```

## Porcentajes: 
En matemática, se denomina porcentaje a una porción proporcional del número 100, por lo tanto, puede expresarse como fracción. Si decimos 50 % (este es el símbolo que representa el porcentaje) significa la mitad de cien; el 100 % es el total. 

Es mucho mas facil imaginarnos la proporcion de una cantidad dentro de otra, si lo pensamos como porcentaje. Por ejemplo:
>Observe la siguiente figura, suponga que el valor representa una magnitud cualquiera, por ejemplo el numero de mujeres en un pueblo, si nos dicen hay 4500 mujeres, es dificil hacernos una idea de que tanta proporcion ocupan respecto a la poblacion total. Pero si nos dicen que en el pueblo el 60% son mujeres, entonces nuestra idea se aclara mucho mas.

En la figura se observa que al deslizar el indicador conforme aumenta el valor respecto al total su porcentaje tambien

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, widgets
from IPython.display import display

cant1 = 5000
#cant2 = 5000

def f(valor, cuadricula):
    porc = valor*100/5000
    g1 = plt.bar(1, cant1, 1)
    g2 = plt.bar(1, valor, 1)
    if cuadricula == 'si':
        plt.grid(True)
    else:
        plt.grid(False)
    
    plt.ylabel('Cantidades')
    plt.xlabel('Concepto')
    plt.xticks(np.arange(0, 2.1, 1),('','Ejemplo', ''))
    plt.yticks(np.arange(0, 5001, 500))
    plt.legend((g1[0], g2[0]), ('Total', 'Valor'))
    plt.title('El porcentaje del valor es {}'.format(porc))
```

```python
if __name__ == "__main__":
    interact(f, valor=widgets.FloatSlider(min=0, max=5000, step=500, value=0),cuadricula=['no','si'])
```

Respuesta

```python
def porcentaje(a,b):
    t = a * 100
    respuesta = t / b
    print(respuesta)
porcentaje(30, 70)
```
