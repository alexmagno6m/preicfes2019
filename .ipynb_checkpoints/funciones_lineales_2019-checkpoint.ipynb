# Funcion lineal

```python
import numpy as np
from bokeh.layouts import column, row
from bokeh.models import CustomJS, ColumnDataSource, Slider, TextInput
from bokeh.plotting import Figure, output_file, show
from bokeh.io import output_notebook, show, reset_output
reset_output()
output_notebook()
x = np.arange(-10,10,0.1)
#x = [x*0.005 for x in range(0, 200)]
y = x

source = ColumnDataSource(data=dict(x=x, y=y))

plot = Figure(plot_width=450, plot_height=450, title="funcion  f(x)=ax+b")
plot.line('x', 'y', source=source, line_width=3, line_alpha=0.6)


plot.xaxis.axis_line_width = 3
plot.xaxis.axis_line_color = "red"
plot.xaxis.fixed_location = 0
plot.xaxis.axis_label = 'x'
plot.yaxis.axis_line_width = 3
plot.yaxis.axis_line_color = "red"
plot.yaxis.fixed_location = 0


intercepto = Slider(start=-10, end=10, value=0, step=0.1, title="intercepto (b)")
pendiente = Slider(start=-5, end=5, value=1, step=0.1, title="pendiente (a, numero que multiplica a x)")

callback = CustomJS(args=dict(source=source, int1=intercepto, pen=pendiente), code="""
    var data = source.data;
    var int1 = int1.value;
    var pen = pen.value;
    var x = data['x']
    var y = data['y']
    for (var i = 0; i < x.length; i++) {
        //y[i] = Math.pow(x[i], f)
        y[i]= x[i]*pen + int1;
    }
    source.change.emit();
""")


intercepto.js_on_change('value', callback)
pendiente.js_on_change('value', callback)

layout = row( plot, column (intercepto, pendiente))

show(layout)
```

# Funciones CuadrÃ¡ticas

```python
x = np.arange(-10,10,0.1)
#x = [x*0.005 for x in range(0, 200)]
y = x**2

source = ColumnDataSource(data=dict(x=x, y=y))

plot = Figure(plot_width=420, plot_height=420, title="funcion  f(x)=ax^2+bx+c")
plot.line('x', 'y', source=source, line_width=3, line_alpha=0.6)


plot.xaxis.axis_line_width = 3
plot.xaxis.axis_line_color = "red"
plot.xaxis.fixed_location = 0
plot.xaxis.axis_label = 'x'
plot.yaxis.axis_line_width = 3
plot.yaxis.axis_line_color = "red"
plot.yaxis.fixed_location = 0
#v = -(b.value/2*a.value)
#vertice = a.value*v**2+b.value*v+c.value


a = Slider(start=-10, end=10, value=1, step=0.1, title="a")
b = Slider(start=-10, end=10, value=0, step=0.1, title="b")
c = Slider(start=-20, end=20, value=0, step=0.1, title="c")
#text_input = TextInput(value="vertice={}".format(vertice), title="Label:")
callback = CustomJS(args=dict(source=source, a1=a, b1=b, c1=c), code="""
    var data = source.data;
    var a1 = a1.value;
    var b1 = b1.value;
    var c1 = c1.value;
    var x = data['x']
    var y = data['y']
    for (var i = 0; i < x.length; i++) {
        y[i] = Math.pow(x[i], 2)*a1 + x[i]*b1 + c1;
        //y[i]= x[i]*pen + int1;
    }
    source.change.emit();
""")


a.js_on_change('value', callback)
b.js_on_change('value', callback)
c.js_on_change('value', callback)


layout = row( plot, column( a, b, c))

show(layout)
```
